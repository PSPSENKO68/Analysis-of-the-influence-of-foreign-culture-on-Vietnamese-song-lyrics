```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### chinese-hanviet-cognates.tsv

```python
# Lá»–I CÅ¨: pd.read_tsv khÃ´ng tá»“n táº¡i
# Sá»¬A Láº I: DÃ¹ng read_csv vá»›i sep='\t'
hv1 = pd.read_csv('chinese-hanviet-cognates.tsv', sep='\t')
rows_initial = len(hv1)

# Lá»c cá»™t, xÃ³a dÃ²ng trá»‘ng
hv1 = hv1[["hanviet"]].dropna()
rows_after_dropna = len(hv1)
# LÃ m sáº¡ch dá»¯ liá»‡u: viáº¿t thÆ°á»ng vÃ  xÃ³a khoáº£ng tráº¯ng thá»«a
hv1["hanviet"] = hv1["hanviet"].str.lower().str.strip()

# ThÃªm cá»™t Ä‘Ã¡nh dáº¥u nguá»“n
hv1["source"] = "ryanphung"

print("ğŸ“Š THá»NG KÃŠ Dá»® LIá»†U")
print(f"ğŸ”¹ Sá»‘ dÃ²ng ban Ä‘áº§u        : {rows_initial}")
print(f"ğŸ”¹ Sau dropna             : {rows_after_dropna}")

print(hv1.head())
```

### hanviet.csv

```python
import pandas as pd

# Äá»c file
hv2 = pd.read_csv("hanviet.csv")

# Sá»‘ dÃ²ng ban Ä‘áº§u (toÃ n bá»™ file)
rows_initial = len(hv2)

# Chá»‰ giá»¯ cá»™t hanviet + xÃ³a NaN
hv2 = hv2[["hanviet"]].dropna()
rows_after_dropna = len(hv2)

# 1. XÃ³a bá» cÃ¡c kÃ½ tá»± thá»«a [ ] ' "
hv2["hanviet"] = hv2["hanviet"].str.replace(r"[\[\]\'\"]", "", regex=True)

# 2. TÃ¡ch chuá»—i thÃ nh danh sÃ¡ch (list) dá»±a trÃªn dáº¥u pháº©y
hv2["hanviet"] = hv2["hanviet"].str.split(",")

# 3. PHÃ‚N RÃƒ (Explode)
hv2 = hv2.explode("hanviet")
rows_after_explode = len(hv2)

# 4. LÃ m sáº¡ch khoáº£ng tráº¯ng thá»«a vÃ  viáº¿t thÆ°á»ng
hv2["hanviet"] = hv2["hanviet"].str.strip().str.lower()

# 5. GÃ¡n nguá»“n
hv2["source"] = "phongp-hanviet-pinyin"

# In thá»‘ng kÃª
print("ğŸ“Š THá»NG KÃŠ Dá»® LIá»†U")
print(f"ğŸ”¹ Sá»‘ dÃ²ng ban Ä‘áº§u        : {rows_initial}")
print(f"ğŸ”¹ Sau dropna             : {rows_after_dropna}")
print(f"ğŸ”¹ Sau explode + clean    : {rows_after_explode}")
print(f"ğŸ”¹ Sá»‘ dÃ²ng cuá»‘i cÃ¹ng      : {len(hv2)}")

print("\nğŸ” SAMPLE Dá»® LIá»†U")
print(hv2.head(50))
```

### Gá»™p vÃ  lÃ m sáº¡ch táº¡o thÆ° viá»‡n tá»« hÃ¡n viá»‡t

```python
import pandas as pd



# Thá»‘ng kÃª ban Ä‘áº§u
rows_hv1 = len(hv1)
rows_hv2 = len(hv2)

# Gá»™p (append / ná»‘i dÃ²ng)
hv_all = pd.concat([hv1, hv2], ignore_index=True)

# XÃ³a rá»—ng
hv_all_cleaned = hv_all.dropna(subset=['hanviet'])

# Tá»•ng sá»‘ dÃ²ng sau khi gá»™p
rows_total = len(hv_all)


# Reset index cho Ä‘áº¹p
hv_all_cleaned = hv_all.reset_index(drop=True)



# LÆ°u file
hv_all_cleaned.to_csv("hanviet_dictionary.csv", index=False, encoding="utf-8")

# In thá»‘ng kÃª
print("ğŸ“Š THá»NG KÃŠ Gá»˜P FILE")
print(f"ğŸ”¹ Sá»‘ dÃ²ng hv1      : {rows_hv1}")
print(f"ğŸ”¹ Sá»‘ dÃ²ng hv2      : {rows_hv2}")
print(f"ğŸ”¹ Tá»•ng sau khi gá»™p : {rows_total}")
```
